Importation des différentes librairies

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

Importation du set de données 'iris.data'

```python
df = pd.read_csv("http://chendeb.free.fr/iris.data", names=["sl","sw","pl", "pw", "species"])
```

```python
df
```

Création du scatter plot de 'sl' et 'sw' avec Matplotlib

```python
sl = df['sl']
sw = df['sw']

plt.scatter(sl, sw)

plt.title('Scatter Plot de sl et sw')
plt.xlabel('Longueur du Sépale (sl)')
plt.ylabel('Largeur du Sépale (sw)')

plt.show()
```

Création d'un FacetGrid Plot avec Seaborn

```python
sl = df['sl']
sw = df['sw']
species = df['species']

g = sns.FacetGrid(df, hue='species', height=5)
g.map(plt.scatter, 'sl', 'sw')

g.set_axis_labels("SepalLengthCm", "SepalWidthCm")
g.add_legend(frameon=True)

plt.show()
```

Création d'un Boxplot avec Seaborn

```python
pl = df['pl']
species = df['species']

sns.boxplot(x=df['species'], y=df['pl']);

plt.ylabel('PetalLengthCm')

plt.show()
```
